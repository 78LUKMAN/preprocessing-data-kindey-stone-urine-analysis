**IMPORT DATA**
- pd.read_csv(): Mengimpor dataset dari file CSV.
- dataset.iloc[:, :-1]: Mengambil semua kolom kecuali kolom terakhir sebagai variabel fitur (X).
- dataset.iloc[:, -1]: Mengambil kolom terakhir sebagai variabel target (Y).

```python
dataset = pd.read_csv('dataset/kindey_stone_urine_analysis_modified.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

**MENGATASI MISSING VALUES**
- SimpleImputer(): Digunakan untuk mengisi missing values. Pada bagian ini, kita mengisi nilai yang hilang pada kolom 1 hingga 5 di variabel X dengan mean.

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 1:6])
x[:, 1:6] = imputer.transform(x[:, 1:6])
```

```python
print(x)
```

Cara yang tepat untuk mengimputasi data target adalah menggunakan metode yang kompatibel dengan LabelEncoder. Saat kita mencoba menggunakan y[:, np.newaxis], ini akan menghasilkan masalah karena y seharusnya tetap berbentuk 1D array (bukan 2D array).

```python
# imputerY = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
# imputerY.fit(y[:, np.newaxis])
# y[:, np.newaxis] = imputerY.transform(y[:, np.newaxis])
# print(y)

y = y.reshape(-1, 1)  # Ubah y menjadi array 2D
imputerY = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
y = imputerY.fit_transform(y).ravel()  # Gunakan ravel() untuk mengubah kembali ke array 1D
print(y)
```

**LABEL ENCODING UNTUK Y**
- LabelEncoder(): Digunakan untuk mengubah variabel kategorikal menjadi angka. Ini akan mengubah no dan yes di variabel target menjadi 0 dan 1.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

**TEST TRAIN SPLIT**
- train_test_split(): Membagi dataset menjadi data latih (train) dan data uji (test). Di sini, kita membagi 80% data sebagai training set dan 20% sebagai testing set.
- random_state memastikan bahwa pembagian data konsisten setiap kali kode dijalankan.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

**STANDARDISASI FITUR**
- StandardScaler(): Digunakan untuk menstandarkan fitur numerik agar memiliki distribusi normal dengan mean 0 dan standard deviation 1. Di sini, kuta melakukan scaling pada kolom ke-1 hingga ke-5.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:, 1:6] = sc.fit_transform(x_train[:, 1:6])
x_test[:, 1:6] = sc.transform(x_test[:, 1:6])
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 1. Inisialisasi model
classifier = LogisticRegression(random_state=0)

# 2. Latih model dengan data training
classifier.fit(x_train, y_train)

# 3. Prediksi dengan data uji
y_pred = classifier.predict(x_test)

```

```python
print("Accuracy Logistic Regression:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
